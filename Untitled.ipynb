```python
def prob_emp(year, age, sex) :
    if sex == 0 :
        if age <= 14 :
            return 0
        elif age > 14 and age <= 19 :
            pemp = float(df['_b_cons'][0:1]) + float(df['_b_time'][0:1]) * year + float(df['_b_time_sq'][0:1]) * (year**2)
            return pemp / 100
        elif age > 19 and age <= 24 :
            pemp = float(df['_b_cons'][1:2]) + float(df['_b_time'][1:2]) * year + float(df['_b_time_sq'][1:2]) * (year**2)
            return pemp / 100
        elif age > 24 and age <= 29 :
            pemp = float(df['_b_cons'][2:3]) + float(df['_b_time'][2:3]) * year + float(df['_b_time_sq'][2:3]) * (year**2)
            return pemp / 100
        elif age > 29 and age <= 34 :
            pemp = float(df['_b_cons'][3:4]) + float(df['_b_time'][3:4]) * year + float(df['_b_time_sq'][3:4]) * (year**2)
            return pemp / 100
        elif age > 34 and age <= 39 :
            pemp = float(df['_b_cons'][4:5]) + float(df['_b_time'][4:5]) * year + float(df['_b_time_sq'][4:5]) * (year**2)
            return pemp / 100
        elif age > 39 and age <= 44 :
            pemp = float(df['_b_cons'][5:6]) + float(df['_b_time'][5:6]) * year + float(df['_b_time_sq'][5:6]) * (year**2)
            return pemp / 100
        elif age > 44 and age <= 49 :
            pemp = float(df['_b_cons'][6:7]) + float(df['_b_time'][6:7]) * year + float(df['_b_time_sq'][6:7]) * (year**2)
            return pemp / 100
        elif age > 49 and age <= 54 :
            pemp = float(df['_b_cons'][7:8]) + float(df['_b_time'][7:8]) * year + float(df['_b_time_sq'][7:8]) * (year**2)
            return pemp / 100
        elif age > 54 and age <= 59 :
            pemp = float(df['_b_cons'][8:9]) + float(df['_b_time'][8:9]) * year + float(df['_b_time_sq'][8:9]) * (year**2)
            return pemp / 100
        elif age > 59 and age <= 64 :
            pemp = float(df['_b_cons'][9:10]) + float(df['_b_time'][9:10]) * year + float(df['_b_time_sq'][9:10]) * (year**2)
            return pemp / 100
        elif age > 64 :
            pemp = float(df['_b_cons'][10:11]) + float(df['_b_time'][10:11]) * year + float(df['_b_time_sq'][10:11]) * (year**2)
            return pemp / 100
        else:
            raise Exception("An error has occured with the age")
    elif sex == 1 :
        if age <= 14 :
            return 0
        elif age > 14 and age <= 19 :
            pemp = float(df['_b_cons'][11:12]) + float(df['_b_time'][11:12]) * year + float(df['_b_time_sq'][11:12]) * (year**2)
            return pemp / 100
        elif age > 19 and age <= 24 :
            pemp = float(df['_b_cons'][12:13]) + float(df['_b_time'][12:13]) * year + float(df['_b_time_sq'][12:13]) * (year**2)
            return pemp / 100
        elif age > 24 and age <= 29 :
            pemp = float(df['_b_cons'][13:14]) + float(df['_b_time'][13:14]) * year + float(df['_b_time_sq'][13:14]) * (year**2)
            return pemp / 100
        elif age > 29 and age <= 34 :
            pemp = float(df['_b_cons'][14:15]) + float(df['_b_time'][14:15]) * year + float(df['_b_time_sq'][14:15]) * (year**2)
            return pemp / 100
        elif age > 34 and age <= 39 :
            pemp = float(df['_b_cons'][15:16]) + float(df['_b_time'][15:16]) * year + float(df['_b_time_sq'][15:16]) * (year**2)
            return pemp / 100
        elif age > 39 and age <= 44 :
            pemp = float(df['_b_cons'][16:17]) + float(df['_b_time'][16:17]) * year + float(df['_b_time_sq'][16:17]) * (year**2)
            return pemp / 100
        elif age > 44 and age <= 49 :
            pemp = float(df['_b_cons'][17:18]) + float(df['_b_time'][17:18]) * year + float(df['_b_time_sq'][17:18]) * (year**2)
            return pemp / 100
        elif age > 49 and age <= 54 :
            pemp = float(df['_b_cons'][18:19]) + float(df['_b_time'][18:19]) * year + float(df['_b_time_sq'][18:19]) * (year**2)
            return pemp / 100
        elif age > 54 and age <= 59 :
            pemp = float(df['_b_cons'][19:20]) + float(df['_b_time'][19:20]) * year + float(df['_b_time_sq'][19:20]) * (year**2)
            return pemp / 100
        elif age > 59 and age <= 64 :
            pemp = float(df['_b_cons'][20:21]) + float(df['_b_time'][20:21]) * year + float(df['_b_time_sq'][20:21]) * (year**2)
            return pemp / 100
        elif age > 64 :
            pemp = float(df['_b_cons'][21:]) + float(df['_b_time'][21:]) * year + float(df['_b_time_sq'][21:]) * (year**2)
            return pemp / 100
        else:
            raise Exception("An error has occured with the age")
    else:
        raise Exception("An error has occured with the sex")
    
```

```python
def exper(age, educ, sex) :
    potential = age - (educ + 4)
    exp = 0
    for yr in range(potential) :
#         print(2011 - potential + yr)
#         print(yr + educ + 4)
        pemp = prob_emp((2011 - potential) + yr , yr + educ + 4 , sex)
#         print(pemp)
        exp += pemp
#         print(exp)
    return exp
# def exper(individual) :
    
#     potential = individual[0:0] - (individual[0:1] + 4)
#     exp = 0
#     for yr in range(potential) :
# #         print(2011 - potential + yr)
# #         print(yr + educ + 4)
#         pemp = prob_emp((2011 - potential) + yr , yr + individual[0:1] + 4 , individual[0:2])
# #         print(pemp)
#         exp += pemp
# #         print(exp)
#     return exp
```
